```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_FWD(i):
    f=open("../dump/FWD_"+str(i)+".txt")
    return [int(x) for x in f.read().split()]
```

```python
data={}
for i in range(1,6):
    data[i]=load_FWD(i)
    
df=pd.DataFrame(data)
```

```python
sns.violinplot(data=df)
```
